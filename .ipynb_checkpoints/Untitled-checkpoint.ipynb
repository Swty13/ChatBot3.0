```python
import pandas as pd
```

```python
df = pd.read_json("data.json")
```

```python
df.iloc[0]
```

```python
rajdata = df.iloc[0][0]
rajdata
```

```python
ids=[]
name=[]
types=[]
ratings=[]
travel_guide=[]
city_statename=[]
city_bestTime=[]
city_reviews=[]
city_details=[]
city_famous_for=[]
city_images=[]
```

```python
furgi = []
for k,v in rajdata[0].items():
    if k=="city_bestTime":
       furgi.append(v)
furgi
```

```python
for i in range(0,len(rajdata)):
    for k,v in rajdata[i].items():  
        if k=="_id":
            ids.append(v)
        if k=="name":
            name.append(v)
        if k=="type":
            types.append(v)
        if k=="ratings":
            ratings.append(v)
        if k=="travel_guide":
            travel_guide.append(v)
        if k=="city_statename":
            city_statename.append(v)
        if k=="city_bestTime":
            city_bestTime.append(v)
        if k=="city_reviews":
            city_reviews.append(v)
        if k=="city_details":
            city_details.append(v)
        if k=="city_famous_for":
            city_famous_for.append(v)
        if k=="city_images":
            city_images.append(v)
```

```python
len(ids)
```

```python
rajdata_content = {
      "id":ids,
      "name":name,
      "type":types,
      "ratings":ratings,
      "travel_guide":travel_guide,
      "city_statename":city_statename,
      "city_bestTime":city_bestTime,
    "city_reviews":city_reviews,
    "city_details":city_details,
    "city_famous_for":city_famous_for,
    "city_images":city_images
}
```

```python
travel_content = pd.DataFrame(rajdata_content)
```

```python
reach = df.iloc[1][0]
```

```python
len(reach)
```

```python
reach_ids = []
reach_city_name = []
reach_By_Bus = []
reach_By_Train = []
reach_By_Flight = []
reach_By_Cab = []
```

```python
for i in range(0,len(reach[0])):
    for k,v in reach[i].items():  
        if k=="_id":
            reach_ids.append(v)
        if k=="city_name":
            reach_city_name.append(v)
        if k=="By Bus":
            reach_By_Bus.append(v)
        if k=="By Train":
            reach_By_Train.append(v)
        if k=="By Flight":
            reach_By_Flight.append(v)
        if k=="By Cab/Taxi":
            reach_By_Cab.append(v)
```

```python
reach_content = {
      "id":reach_ids,
      "reach_city_name":reach_city_name,
      "reach_By_Bus":reach_By_Bus,
      "reach_By_Train":reach_By_Train,
      "reach_By_Flight":reach_By_Flight,
      "reach_By_Cab":reach_By_Cab,
}
```
